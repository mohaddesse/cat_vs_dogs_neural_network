## create labels and data from this images

```python
data=[]
path_train="./dogs-vs-cats/train/"
images_train=os.listdir(path_train)
for img in images_train[:500]:
    img_path=os.path.join(path_train,img)
    if img.split(".")[0]=="dog":
        img_dog=cv2.imread(img_path)
        img_resize=cv2.resize(img_dog,(64,64))
        data.append([img_resize,1])
    else:
        img_cat=cv2.imread(img_path)
        img_resize=cv2.resize(img_cat,(64,64))
        data.append([img_resize,0])

len(data)        
    
```

```python
plt.imshow(data[100][0])
```

```python
x=[]
y=[]
for arr in data:
    x.append(arr[0])
    y.append(arr[1])
x=np.array(x)
y=np.array(y)
(x.shape,y.shape)
```
